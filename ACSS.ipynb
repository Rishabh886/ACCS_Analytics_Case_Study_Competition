```python
data1 = pd.read_excel('ACCS Data Set.xlsx',sheet_name='payment volume')
data2 = pd.read_excel('ACCS Data Set.xlsx',sheet_name='payment method')
data3 = pd.read_excel('ACCS Data Set.xlsx',sheet_name='payment platform')
data_4 = pd.read_excel('ACCS Data Set.xlsx',sheet_name='traffic')
```

```python
# Converting type of date from object to date type
data1['Date'] = pd.to_datetime(data1['Date'])
data2['Date'] = pd.to_datetime(data2['Date'])
data3['Date'] = pd.to_datetime(data3['Date'])
data_4['Day'] = pd.to_datetime(data_4['Day'])
```

```python
data1
```

```python
# Function for extracting year, month, weekday and day from given date
def date_extract(data):
    data['year'] = data['Date'].dt.year
    data['month'] = data['Date'].dt.month
    data['weekday'] = data['Date'].dt.weekday
    data['day_no'] = data['Date'].dt.day
```

```python
# Restricting total rows to 345
data4 = data_4.iloc[:345,:]
```

```python
# Revering whole dataset to match date column with other tables 
data4 = data4.iloc[::-1]
```

```python
data4.rename(columns={'Day':'Date'},inplace=True)
```

```python
# Merging data1 with data4
df14 = pd.merge(data1, data4, how='outer')
df14
```

```python
# Computing metrics such as Pages per session and Sessions per unique visitor
df14['Pages per sessions'] = df14['Page views']/df14['Site sessions']
df14['Sessions per unique visitor'] = df14['Site sessions']/df14['Unique visitors']
```

```python
df14
```

```python
# Splitting average session duration to minutes and seconds
df14['Session_dur_min'] = (df14['Avg. session duration'].str.split(' ').str[0].str[0])
df14['Session_dur_sec'] = (df14['Avg. session duration'].str.split(' ').str[1].str.rstrip('s'))
df14['Session_dur_min'] = pd.to_numeric(df14['Session_dur_min'])
df14['Session_dur_sec'] = pd.to_numeric(df14['Session_dur_sec'])
```

```python
# Making a new column total session duration containing avg session duration in seconds
df14['Tot_session_dur'] = df14['Session_dur_min']*60 + df14['Session_dur_sec']
```
